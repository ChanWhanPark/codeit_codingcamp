### 산점도

```python
df.plot(kind='scatter', x='math score', y='reading score')
```

```python
df.plot(kind='scatter', x='math score', y='writing score')
```

```python
df.plot(kind='scatter', x='reading score', y='writing score')
```

```python
import numpy as np
import pylab
# 산점도에서 추세선 그리기

x = df['math score']
y = df['reading score']

z = np.polyfit(x, y, 1)
p = np.poly1d(z)

pylab.plot(x, y, 'o')
pylab.plot(x, p(x), 'r--')

print( 'y=%.6fx+(%.6f)' % (z[0], z[1]))
```
