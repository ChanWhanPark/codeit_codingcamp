#### 카드사 고객 분석

데이터의 중요성을 깨달은 “삼송카드”와 “현디카드”가 협업을 하기로 결정했습니다.

두 카드사는 사람들이 요일별로 지출하는 평균 금액을 “요일”, “식비", “교통비”, “문화생활비”, “기타” 카테고리로 정리해서 우리에게 공유해 주기로 했는데요. 각각 samsong.csv 파일과 hyundee.csv 파일을 보냈습니다.

두 회사의 데이터를 활용해서, 사람들의 요일별 문화생활비를 분석해보려 합니다. 아래와 같은 형태로 출력이 되도록 DataFrame을 만들어보세요.

```python
import pandas as pd

samsong_df = pd.read_csv('samsong.csv')
hyundee_df = pd.read_csv('hyundee.csv')

# 코드를 작성하세요.

day_series = samsong_df.loc[:, '요일']
sam_series = samsong_df.loc[:, '문화생활비']
hyun_series = hyundee_df.loc[:, '문화생활비']

new_df = {
    'day' : day_series,
    'samsong' : sam_series,
    'hyundee' : hyun_series
}

df = pd.DataFrame(new_df)
df
```
